### Variavel Quantidade_transa√ß√µes

```python
c = ['Quantidade_Transacoes', 'Status_Cliente']
qdt_transacoes = df_num[c].copy()
qdt_transacoes.head()
```

```python
fig_barplot_quantidade_Transacoes = catchPath('image/barplot_quantidade_Transacoes.png')
fig_barplot_quantidade_Transacoes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cores_status = ["#219ebc", "#023047"]

plt.figure(figsize=(12, 6))

sns.boxplot(
    data=qdt_transacoes, 
    x="Status_Cliente", 
    y="Quantidade_Transacoes",
    hue="Status_Cliente",     
    palette=cores_status,      
    legend=False               
)


plt.title("Distribui√ß√£o da Quantidade de Transa√ß√µes por Status do Cliente", fontsize=16)
plt.xlabel("Status do Cliente", fontsize=12)
plt.ylabel("Quantidade de Transa√ß√µes", fontsize=12)



plt.savefig(fig_barplot_quantidade_Transacoes, dpi=300, bbox_inches='tight')
plt.show()
```

```python

mediana_qde_transacoes = qdt_transacoes.groupby("Status_Cliente")["Quantidade_Transacoes"].median().reset_index()
mediana_qde_transacoes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Defina a lista de cores
minha_paleta = ['#219ebc', '#023047']

plt.figure(figsize=(8, 5))

sns.barplot(
    data=mediana_qde_transacoes, 
    x="Status_Cliente", 
    y="Quantidade_Transacoes",
    hue="Status_Cliente", 
    palette=minha_paleta,
    legend=False
)

plt.title("Mediana de Transa√ß√µes por Status")
plt.tight_layout()
plt.show()
```
